```python
import pandas as pd

# Cargar el dataset
data = pd.read_csv('ds2.csv', delimiter=';', encoding='utf-8')

# Mostrar información general del dataset
print("Información del dataset:")
print(data.info())

# Descripción estadística de las columnas numéricas y categóricas
print("\nDescripción estadística de las columnas numéricas:")
print(data.describe())

print("\nDescripción estadística de las columnas categóricas:")
print(data.describe(include=['object']))

# Verificar la cantidad de valores nulos por columna
print("\nValores nulos por columna:")
print(data.isnull().sum())

# Mostrar la cantidad de valores únicos por columna
print("\nValores únicos por columna:")
for column in data.columns:
    print(f"{column}: {data[column].nunique()}")

# Análisis específico de algunas columnas clave
print("\nDistribución de la columna 'NOMBRE_TDOC':")
print(data['NOMBRE_TDOC'].value_counts())

print("\nDistribución de la columna 'NOMBRE_SDOC':")
print(data['NOMBRE_SDOC'].value_counts())

print("\nDistribución de la columna 'FORMA_ENTRADA_PREL':")
print(data['FORMA_ENTRADA_PREL'].value_counts())

print("\nDistribución de la columna 'NOMBRE_OFIC':")
print(data['NOMBRE_OFIC'].value_counts())

# Mostrar algunas filas del dataset para verificar el contenido
print("\nPrimeras 5 filas del dataset:")
print(data.head())

# Analizar correlación entre columnas numéricas (si es relevante)
numerical_data = data.select_dtypes(include=['number'])
print("\nMatriz de correlación entre columnas numéricas:")
print(numerical_data.corr())

# Longitud de los textos en la columna 'TEXTO_SOLICITUD_PREL'
print("\nLongitud de los textos en 'TEXTO_SOLICITUD_PREL':")
data['longitud_texto'] = data['TEXTO_SOLICITUD_PREL'].fillna('').apply(len)
print(data['longitud_texto'].describe())

# Análisis de fechas
print("\nAnálisis de la columna 'FECHA_ALTA_PREL':")
data['FECHA_ALTA_PREL'] = pd.to_datetime(data['FECHA_ALTA_PREL'], errors='coerce')
print(data['FECHA_ALTA_PREL'].describe())
print("\nDistribución de las fechas en 'FECHA_ALTA_PREL':")
print(data['FECHA_ALTA_PREL'].value_counts().sort_index())
```
